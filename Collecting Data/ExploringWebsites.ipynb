### **`National Labor Exchange`**

#### API requests

```python
# Sending an API request
import requests

headers = {
    'authority': 'prod-search-api.jobsyn.org',
    'accept': 'application/json',
    'accept-language': 'ar,en-US;q=0.9,en;q=0.8,fr;q=0.7,fr-FR;q=0.6,en-GB;q=0.5',
    'origin': 'https://usnlx.com',
    'referer': 'https://usnlx.com/',
    'sec-ch-ua': '"Chromium";v="118", "Microsoft Edge";v="118", "Not=A?Brand";v="99"',
    'sec-ch-ua-mobile': '?0',
    'sec-ch-ua-platform': '"Windows"',
    'sec-fetch-dest': 'empty',
    'sec-fetch-mode': 'cors',
    'sec-fetch-site': 'cross-site',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36 Edg/118.0.2088.46',
    'x-origin': 'usnlx.com',
}

params = {
    'page': '1',
    'offset': '15',
    'num_items': '15',
}

response = requests.get('https://prod-search-api.jobsyn.org/api/v1/solr/search', params=params, headers=headers)
response
```

```python
# Reading json data
data = response.json()
data
```

```python
# Investigating the data
print(data['pagination']['page_size']) # The chosed page size
print(data['pagination']['offset']) # The Chosed offset
print(data['pagination']['page']) # The Current page
print(data['pagination']['total_pages']) # Total pages for the chosen pagesize
print(data['pagination']['total']) # Total job posts
```

```python
# Job details
for info in data['jobs'][0]:
    print(f"{info}: {data['jobs'][0][info]}")
    print('='*30)
```

#### Wrap up

Very well. Scraping data from the *National Labor Exchange* web site is very easy. The web site provide a Free API sevice though, but even with that I have been able to scrape 20000 job post data at once with a `simble HTTP GET request`.

### **`Dice`**

#### API requests

After some expermenting, I found that this web site don't use AJAX Requests to retrive data. So we have to take the second approach and scrape the data from the html, page after page, and job after job.

#### BeautifulSoup

```python
import requests
from bs4 import BeautifulSoup

responce = requests.get("https://www.dice.com/jobs/q--jobs?p=1")
responce
```

```python
data = responce.text
data
```

```python
soup = BeautifulSoup(data, 'html.parser')
soup
```

```python
# Exploring the page
data = soup.find('div', class_="col-md-12 col-lg-12")
offset = int(data.find('span').text.split(' ')[-1])
print('offset:', offset)
data = soup.find('span', id="posiCountId")
allPositions = int(data.text.replace(',',''))
print('All Positions:', allPositions)
allPages = int(allPositions/offset+1)
print('All pages:', allPages)
```

```python
# get each job link (20 jobs)
data = soup.find_all('a', class_="dice-btn-link loggedInVisited easy-apply")
print(len(data))
for datum in data:
    print(datum['href'])
```
